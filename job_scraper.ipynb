```python
import requests
from bs4 import BeautifulSoup

url = "https://realpython.github.io/fake-jobs/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
job_cards = soup.find_all("div", class_="card-content")

print("Jobs found:", len(job_cards))
```

```python
import csv

with open("jobs.csv", mode="w", newline="", encoding="utf-8-sig") as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Company", "Location", "Link"])

    for job in job_cards:
        title = job.find("h2", class_="title").text.strip()
        company = job.find("h3", class_="company").text.strip()
        location = job.find("p", class_="location").text.strip()
        link = job.find_all("a")[1]["href"]

        writer.writerow([title, company, location, link])

print("CSV saved with data!")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://realpython.github.io/fake-jobs/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")
job_cards = soup.find_all("div", class_="card-content")

print("Jobs found:", len(job_cards))

with open("my_jobs_output.csv", mode="w", newline="", encoding="utf-8-sig") as file:
    writer = csv.writer(file)
    writer.writerow(["Title", "Company", "Location", "Link"])

    for job in job_cards:
        title = job.find("h2", class_="title").text.strip()
        company = job.find("h3", class_="company").text.strip()
        location = job.find("p", class_="location").text.strip()
        link = job.find_all("a")[1]["href"]

        writer.writerow([title, company, location, link])

print("Done.")
```
